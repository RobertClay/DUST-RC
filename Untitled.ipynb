testing for Unscented Kalman Filter using circular motion about a region filled with landmarks.

First need to define state space looking at the x and y direction positions and velocities as well as the orientation clockwise about east.

$$ state = [x,\dot{x},y,\dot{y},\theta]$$

Typical state transition matrix for velocities. for unscented need to expand this to non linear function

$$ F = \begin{bmatrix}
{{1,dt,0,0}\\{0,1,0,0}\\{0,0,1,dt}\\{0,0,0,1}
}
\end{bmatrix}$$



```python
seed(10)
count=50
dt =1
x_var = 0.125
y_var = 0.125
r_var = 0.5
landmarks = array([[5,-5],[5,5],[-5,-5],[-5,5]])   
radius = 4
y_lim = [-5,5]
x_lim = [-5,5] 
beta = 2*pi/count
vel = (2*radius*pi/(count*dt),2*radius*pi/(count*dt)) #determined by above to give 1 revolution and nice looking plots
pos=(radius,0,pi/2)
```

```python
class SLAM_PINGS(object):
    def __init__(self, pos=(0, 0,0), vel=(0,0), 
                 count=0,dt =0.1,
                 x_var = 0.5, y_var = 0.5,
                 landmarks = array([0,0]),
                 radius=1):
        self.dt = dt
        self.vel = vel
        self.pos = [pos[0], pos[1]]
        self.times = arange(0,count,dt)
        self.theta_x = vel[0]/radius
        self.theta_y = vel[1]/radius

        #set up placeholder matrices and IC rows

        self.positions = zeros((count+1,2)) 
        self.positions[0,:] = array([pos])# start point
        self.landmark_distances = zeros((count+1,len(landmarks)))
        
        for j in range(len(landmarks)):
            self.landmark_distances[0,(j)] = norm(self.positions[0,:]-landmarks[j,:])
            #self.landmark_distances[0,(2*j)+1] = (positions[0,:]-landmarks[j,:])[1]
                
            
            
    def write(self):
        positions = self.positions
        landmark_distances = self.landmark_distances
        dt = self.dt
        vel = self.vel
        theta_x = self.theta_x
        theta_y= self.theta_y

        
        for i in range(positions.shape[0]-1):
            positions[i+1,0] =  radius*cos(dt*i*theta_x) + x_var*randn()
            if positions[i+1,0] <x_lim[0]:
                positions[i+1,0] = x_lim[0]
            elif positions[i+1,0] >x_lim[1]:
                positions[i+1,0] = x_lim[1]
            positions[i+1,1] =  radius*sin( dt*i*theta_y)   + y_var*randn()
            if positions[i+1,1] <y_lim[0]:
                positions[i+1,1] = y_lim[0]
            elif positions[i+1,1] > y_lim[1]:
                positions[i+1,1] = y_lim[1]
            
            for j in range(len(landmarks)):
                landmark_distances[i+1,j] = norm(landmarks[j,:]-positions[i+1,:]) + r_var*randn()
                #landmark_distances[i+1,(2*j)+1] = (landmark\s[j,:]-positions[i+1,:])[1]


        
        return positions,landmark_distances
    
sp = SLAM_PINGS(pos,vel,count,dt,x_var,y_var,landmarks,radius)
positions,landmark_distances = sp.write()      

plt.plot(positions[:,0],positions[:,1],label = "actual")
#for j in range(landmark_distances.shape[1]):
#    plt.plot(landmark_distances[:,j],label = f"track {j}")
plt.scatter(landmarks[:,0],landmarks[:,1])
plt.legend()

plt.figure()
labels = ["bottom right","top right","bottom left","top left"]
for j in range(landmark_distances.shape[1]):   
    plt.plot(landmark_distances[:,j],label = labels[j])
plt.legend()
```

```python
class SLAM_PINGS(object):
    def __init__(self, pos=(0, 0,0), vel=(0,0), 
                 count=0,dt =0.1,
                 x_var = 0.5, y_var = 0.5,
                 landmarks = array([0,0]),
                 radius=1,alpha=0):
        self.dt = dt
        self.vel = vel
        self.pos = [pos[0], pos[1],pos[2]]
        self.times = arange(0,count,dt)
        self.theta_x = vel[0]/radius
        self.theta_y = vel[1]/radius

        #set up placeholder matrices and IC rows

        self.positions = zeros((count+1,len(pos))) 
        self.positions[0,:] = array([pos])# start point
        self.landmark_distances = zeros((count+1,len(landmarks)))
        
        for j in range(len(landmarks)):
            self.landmark_distances[0,j] = norm(self.positions[0,:2]-landmarks[j,:])
            #self.landmark_distances[0,(2*j)+1] = (positions[0,:]-landmarks[j,:])[1]
                
            
            
    def write(self):
        positions = self.positions
        landmark_distances = self.landmark_distances
        dt = self.dt
        vel = self.vel
        theta_x = self.theta_x
        theta_y= self.theta_y

        
        for i in range(positions.shape[0]-1):
            
            
            positions[i+1,2] = (positions[i+1,1]+beta)%(2*pi)
            
            positions[i+1,0] =  positions[i+1,0] + radius*(sin(positions[i+1,2]+beta)-sin(positions[i+1,2])) + x_var*randn()
            if positions[i+1,0] <x_lim[0]:
                positions[i+1,0] = x_lim[0]
            elif positions[i+1,0] >x_lim[1]:
                positions[i+1,0] = x_lim[1]
            positions[i+1,1] =   positions[i+1,1] + radius*(-cos(positions[i+1,2]+beta)+cos(positions[i+1,2]))+ y_var*randn()
            if positions[i+1,1] <y_lim[0]:
                positions[i+1,1] = y_lim[0]
            elif positions[i+1,1] > y_lim[1]:
                positions[i+1,1] = y_lim[1]
            
            for j in range(len(landmarks)):
                landmark_distances[i+1,j] = norm(landmarks[j,:]-positions[i+1,:2]) + r_var*randn()
                #landmark_distances[i+1,(2*j)+1] = (landmark\s[j,:]-positions[i+1,:])[1]


        
        return positions,landmark_distances
    
sp = SLAM_PINGS(pos,vel,count,dt,x_var,y_var,landmarks,radius,beta)
positions,landmark_distances = sp.write()      

plt.plot(positions[:,0],positions[:,1],label = "actual")
#for j in range(landmark_distances.shape[1]):
#    plt.plot(landmark_distances[:,j],label = f"track {j}")
plt.scatter(landmarks[:,0],landmarks[:,1])
plt.legend()

plt.figure()
labels = ["bottom right","top right","bottom left","top left"]
for j in range(landmark_distances.shape[1]):   
    plt.plot(landmark_distances[:,j],label = labels[j])
plt.legend()
```

```python
"state transition function"
def f_circ(count,radius,i):
    x_pred = radius*cos(dt*i*theta_y)
    y_pred = radius*cos(dt*i*theta_y)
    xv_pred = (x_pred-x)/dt
    yv_pred = (y_pred-y)/dt
    return x_pred,y_pred,xv_pred,yv_pred
    
    
"measure function"

def h_circ(x,lds,landmarks):
    "turns row of distances from each landmark to approximate location in estimate of position"
    loc = lz.Project(mode= "2D",solver="LSE")

    for i in range(len(landmarks)):
        loc.add_anchor(str(i),landmarks[i,:])

    t,label = loc.add_target()

    for i in range(len(landmarks)):
        #for j in range(len)
        t.add_measure(str(i),x[i])

    loc.solve()
    return t.loc

    
    
    
"UKF function"

def UKF_slampings(lds,dt,radius):
    points = MerweScaledSigmaPoints(n=4,alpha = 0.00001,beta=2,kappa=0)
    ukf = UKF(dim_x = 4,dt=dt,dim_z = len(landmarks), fx = f_circ,hx = h_circ,
             points=points)
    #initiate ukf paramters
    ukf.x = np.array([pos,vel]).flatten()
    ukf.P = np.eye(4)
    ukf.R = np.diag([r_var])
    ukf.Q[0:2,0:2] = Q_discrete_white_noise(2,dt=dt,var=0.1)
    ukf.Q[2:,2:] = Q_discrete_white_noise(2,dt=dt,var=0.1)
    
    track = []
    
    for j in range(count+1):
        ukf.predict(j,radius,j+1)
        ukf.update(lds[j+1,:],landmarks)
    
    return ukf

ukf = UKF_slampings(landmark_distances,dt,radius)
```
