Standard Michaelis Mertens System

$$ X + Y <=>[k1][kn1] Z  $$

\begin{align}
\dot{x} &= -2k_1xy + k_{-1}z\\
\dot{y} &= -2k_1xy + k_{-1}z \\
\dot{z} &= 2k_1xy - k_{-1}z\\
\end{align}


```python
import numpy as np
concs = [.5,.5,0,0,0,0]
k1 = 0.5
kn1 = 0.05
count = 500
var = 0.001
n = len(concs)
c = np.zeros((count,n))
c[0,:] = concs



np.random.seed(10)


for i in range(count-1):
    x = c[i,0]
    y = c[i,1]
    z = c[i,2]
    change = np.array([(-k1*x*y + kn1*z),(-k1*x*y + kn1*z),-(-k1*x*y + kn1*z)]) + np.random.randn(3)*var 

    
    c[i+1,:3]=c[i,:3] +change
    c[i+1,3:] = change


#print(c)


import matplotlib.pyplot as plt
labels = ["x","y","z"]
plt.figure()
for j in range(3):
    plt.plot(c[:,j],label = labels[j])
plt.legend()
plt.figure()
for j in range(3):
    plt.plot(c[:,j+3],label = labels[j])
plt.legend()



```

Require, F,Q,R,P,x


```python
"transition function F"
"predict next row using current row of concentrations"

def Fx(x,k1,kn1):
    
    x1 = x[0]
    y = x[1]
    z = x[2]
    
    change = np.array([(-k1*x1*y + kn1*z),(-k1*x1*y + kn1*z),-(-k1*x1*y + kn1*z)]) 

    x[:3]=+change
    x[3:] = change

    return x


def Hx(row):
    h = row[:3]
    return h
    
    
    
```

```python
from filterpy.kalman import UnscentedKalmanFilter
from filterpy.common import Q_discrete_white_noise
from filterpy.kalman import unscented_transform, MerweScaledSigmaPoints
import scipy.stats as stats


points = MerweScaledSigmaPoints(n=n, alpha=.0000001, beta=2., kappa=.1)


ukf = UnscentedKalmanFilter(dim_x=n, dim_z=3, dt=1., hx=Hx, fx=Fx, points=points)
ukf.x = concs
ukf.P *= np.eye(1)
ukf.R *= np.eye(1)
ukf.Q = np.eye(1)

track = np.zeros((count,n))
track[0,:] = concs
kf_track = np.zeros((count,n))

print(ukf.x)
```

```python
for j in range(count-1):
    track[j+1,:]  = Fx(track[j,:],k1,kn1)
    ukf.predict(kn1 = kn1)
    ukf.update(z = c[j+1,:3])
    kf_track[j+1,:] = ukf.x
```

```python
var_label = ["x","y","z"]

for j in range(c.shape[1]):
    plt.figure()
    plt.plot(kf_track[:,j],label = var_label[j]+" KF" )
    plt.plot(track[:,j],label = var_label[j]+" measurements")
    plt.plot(c[:,j],label = var_label[j]+" Prediction")

    plt.legend()
```
